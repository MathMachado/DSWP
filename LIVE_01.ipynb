### Calculando da folha e encargos, supondo 100% de desoneração da folha

```python
def real_br(my_value):
    a = '{:,.2f}'.format(float(my_value))
    b = a.replace(',','v')
    c = b.replace('.',',')
    return c.replace('v','.')

def calc_inss(Comp, Valor):
    
    if Comp >= 202001 and Comp <= 202002:

        LimInf = 1830.29
        LimSup = 6101.06
        Aliq = 8.0

        if Valor > LimInf:  Aliq = 9
        if Valor > 3050.52: Aliq = 11
        if Valor > LimSup: Valor = LimSup

    if Comp >= 202003 and Comp <= 202012:

        LimInf = 1045
        LimSup = 6101.06
        Aliq = 7.65

        if Valor > LimInf:  Aliq = 9
        if Valor > 2089.60: Aliq = 12
        if Valor > 3134.40: Aliq = 14
        if Valor > LimSup: Valor = LimSup

            
    return round((Valor * Aliq / 100),2)

#Aliquota aplicadas

Aliq_Terceiros = 0.058
Aliq_Rat       = 0.02
Aliq_Fap       = 0.0091
Aliq_Fgts      = 0.08
Aliq_Irpj      = 0.15
Aliq_Adic_Irpj = 0.10
Aliq_Csll      = 0.09

#Cálculo dos Tributos e Encargos

#df_fp['Desc_Seg'] = df_fp['Vl_base_ps'].apply(calc_inss)

df_fp['Desc_Seg'] = df_fp.apply(lambda x: calc_inss(x['Dt_comp'],x['Vl_base_ps']) , axis=1)

df_fp['Sal_Liq']  = df_fp['Vl_base_ps'] - df_fp['Desc_Seg']
Tot_Desc_Seg      = df_fp['Desc_Seg'].sum()
Tot_Base_Ps       = df_fp['Vl_base_ps'].sum()
Tot_Cont_Terc     = Tot_Base_Ps * Aliq_Terceiros
Tot_Gilrat        = Tot_Base_Ps * Aliq_Rat * Aliq_Fap
Tot_Sal_Liq       = Tot_Base_Ps - Tot_Desc_Seg
Tot_Fgts          = Tot_Base_Ps * Aliq_Fgts
Tot_Inss          = Tot_Cont_Terc + Tot_Gilrat
Tot_Enc           = Tot_Inss + Tot_Fgts
Tot_Sal_Enc       = Tot_Base_Ps+Tot_Enc 

print(f'Desc.Seg: {real_br(Tot_Desc_Seg)}')
print(f'Salários: {real_br(Tot_Base_Ps)}')
print(f'Sal.Liq.: {real_br(Tot_Sal_Liq)}')
print(f'Enc.INSS: {real_br(Tot_Inss)}')
print(f'Enc.Fgts: {real_br(Tot_Fgts)}')
print(f'Encargos: {real_br(Tot_Enc)}')
print(f'Sal+Enc : {real_br(Tot_Sal_Enc)}')
```

### Analisando a mesma empresa porém simulando parte da remuneração paga via PLR

### Premissas:

1) Quem ganha menos que 2 salários mínimos receberá o mesmo salário acrescido de um PLR
    no valor de 1 salário

2) Quem ganha mais entr 2 e 5 salários mínimos foi contratado por um salário equivalente a 85% do 
   salário pago pelo mesmo cargo pela Empresa 1, porém receberá um PLR no valor de 4 salários

3) Quem ganha acima de 5 salários mínimos foi contratado por um salário equivalente a 80% do 
   salário pago pelo mesmo cargo pela Empresa 1, porém receberá um PLR no valor de 5 salários

```python
df_fp2 = df_fp.copy()

# IMPLEMENTANDO AS REGRAS:

LimInf = 1045 #salário mínimo 2020

#Ajustando a reduçao dos salários

df_fp2['Vl_base_ps2'] = df_fp2.apply(lambda x: x['Vl_base_ps'] * 1   if x['Vl_base_ps'] < (2 * LimInf) else
                                               x['Vl_base_ps'] * .85 if x['Vl_base_ps'] < (5 * LimInf) else
                                               x['Vl_base_ps'] * .80, axis=1)
# Parcela do PLR paga em junho

df_fp2['Vl_Plr_Parc1'] = df_fp2.apply(lambda x: x['Vl_base_ps2'] * 0.5 * 2 if x['Vl_base_ps'] < (2 * LimInf) and x['Dt_comp'] == 202006 else
                                                x['Vl_base_ps2'] * 0.5 * 4 if x['Vl_base_ps'] < (5 * LimInf) and x['Dt_comp'] == 202006 else
                                                x['Vl_base_ps2'] * 0.5 * 6 if x['Dt_comp'] == 202006 else 0, axis=1)
# Parcela do PLR paga em dezembro

df_fp2['Vl_Plr_Parc2'] = df_fp2.apply(lambda x: x['Vl_base_ps2'] * 0.5 * 2 if x['Vl_base_ps'] < (2 * LimInf) and x['Dt_comp'] == 202012 else
                                                x['Vl_base_ps2'] * 0.5 * 4 if x['Vl_base_ps'] < (5 * LimInf) and x['Dt_comp'] == 202012 else
                                                x['Vl_base_ps2'] * 0.5 * 6 if x['Dt_comp'] == 202012 else 0, axis=1)

#Cálculo dos Tributos e Encargos

df_fp['Desc_Seg']  = df_fp2.apply(lambda x: calc_inss(x['Dt_comp'],x['Vl_base_ps2']) , axis=1)
Tot_Desc_Seg2      = df_fp2['Desc_Seg'].sum()
Tot_Base_Ps2       = df_fp2['Vl_base_ps2'].sum()

Tot_Base_Plr2      = df_fp2['Vl_Plr_Parc1'].sum() + df_fp2['Vl_Plr_Parc2'].sum()

df_fp2['Sal_Liq']  = df_fp2.apply(lambda x: x['Vl_base_ps2']  + 
                                            x['Vl_Plr_Parc1'] + 
                                            x['Vl_Plr_Parc2'] - 
                                            x['Desc_Seg'], axis=1)

Tot_Cont_Terc2     = Tot_Base_Ps2 * Aliq_Terceiros

Tot_Gilrat2        = Tot_Base_Ps2 * Aliq_Rat * Aliq_Fap
Tot_Sal_Liq2       = Tot_Base_Ps2 - Tot_Desc_Seg2

Tot_Fgts2          = Tot_Base_Ps2 * Aliq_Fgts
Tot_Inss2          = Tot_Cont_Terc2 + Tot_Gilrat2
Tot_Enc2           = Tot_Inss2 + Tot_Fgts2 

Tot_Recup_Ir       = Tot_Base_Plr2 * (Aliq_Irpj + Aliq_Adic_Irpj + Aliq_Csll)

Tot_Sal_Plr        = Tot_Base_Ps2+Tot_Base_Plr2
Tot_Sal_Liq_Plr    = Tot_Sal_Liq2+Tot_Base_Plr2
Tot_Sal_Plr_Enc    = Tot_Sal_Plr + Tot_Enc2
Tot_Sal_Plr_Enc_Rec= Tot_Sal_Plr_Enc - Tot_Recup_Ir

print(f'Desc.Seg    : {real_br(Tot_Desc_Seg2)}')
print(f'Salários    : {real_br(Tot_Base_Ps2)}')
print(f'PLR Tot     : {real_br(Tot_Base_Plr2)}')
print(f'Salários+PLR: {real_br(Tot_Sal_Plr)}')
print(f'Sal.Liq.+PLR: {real_br(Tot_Sal_Liq_Plr)}')
print(f'Enc.Fgts    : {real_br(Tot_Fgts2)}')
print(f'Enc.Inss    : {real_br(Tot_Inss2)}')
print(f'Encargos    : {real_br(Tot_Enc2)}')
print(f'Sal+Plr+Enc : {real_br(Tot_Sal_Plr_Enc)}')
print(f'Recup. IRPJ : {real_br(Tot_Recup_Ir)}')
print()
print(f'Sal+Plr+Enc-Recup.IRPJ => {real_br(Tot_Sal_Plr_Enc_Rec)}')
```
