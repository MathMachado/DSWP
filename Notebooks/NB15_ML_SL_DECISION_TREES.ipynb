# **DECISION TREE**

- O que são Árvores de Decisão?
    * Definição: Uma árvore de decisão é um modelo que divide os dados em subconjuntos baseados em condições, organizados em uma estrutura de árvore.

* Objetivo:
    * Classificação: Prever classes categóricas (Ex.: spam vs. não-spam).
    * Regressão: Prever valores contínuos (Ex.: preços de casas).


> Decision Trees possuem estrutura em forma de árvores.

![DecisionTree](https://github.com/MathMachado/Materials/blob/master/DecisionTree.PNG?raw=true)

Fonte: [Um tutorial completo sobre modelagem baseada em árvores de decisão (códigos R e Python)](https://www.vooo.pro/insights/um-tutorial-completo-sobre-a-modelagem-baseada-em-tree-arvore-do-zero-em-r-python/)



* **Principais Vantagens**:
    * São algoritmos fáceis de entender, visualizar e interpretar;
    * Captura facilmente padrões não-lineares presentes nos dados;
    * Requer pouco poder computacional --> Treinar Decision Trees não requer tanto recurso computacional!
    * Lida bem com COLUNAS numéricas ou categóricas;
    * Não requer os dados sejam normalizados;
    * Pode ser utilizado como Feature Engineering ao lidar com Missing Values;
    * Pode ser utilizado como Feature Selection;
    * Não requer suposições sobre a distribuição dos dados por causa da natureza não-paramétrica do algoritmo

* **Principais desvantagens**
    * Propenso a Overfitting, pois Decision Trees podem construir árvores complexas que não sejam capazes de generalizar bem os dados. As coisas complicam muito se a amostra de treinamento possuir outliers. Portanto, **recomenda-se fortemente a tratar os outliers previamente**.
    * Pode criar árvores viesadas se tivermos um dataframe não-balanceado ou que alguma classe seja dominante. Por conta disso, **recomenda-se balancear o dataframe previamente para se evitar esse problema**.

* **Principais Hiperparâmetros**
    * **Gini Index** - é uma métrica que mede a frequência com que um ponto/observação aleatoriamente selecionado seria incorretamente identificado.
        * Portanto, quanto menor o valor de Gini Index, melhor a COLUNA;
    * **Entropy** - é uma métrica que mede aleatoriedade da informação presente nos dados.
        * Portanto, quanto maior a entropia da COLUNA, pior ela se torna para nos ajudar a tomar uma conclusão (classificar, por exemplo).

```
from sklearn.datasets import load_iris # Import the load_iris function from sklearn.datasets module

iris = load_iris()
iris.target
```

```
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Carregar dataset
iris = load_iris()
X, y = iris.data, iris.target

# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e treinar o modelo
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X_train, y_train)

# Previsão e avaliação
y_pred = clf.predict(X_test)
print(f"Acurácia: {accuracy_score(y_test, y_pred):.2f}")

# Visualizar a árvore
plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
```

### **Entendendo o Índice de Gini**

O **Índice de Gini** é uma métrica usada em árvores de decisão para medir a **impureza** de um nó. Ele ajuda a determinar a melhor divisão em cada estágio da construção da árvore. O objetivo é escolher a divisão que **minimiza a impureza**, criando nós que são o mais homogêneos possível (ou seja, com maior proporção de uma única classe).

---

### **Fórmula do Índice de Gini**

A fórmula para calcular o Índice de Gini em um nó é:

$𝐺𝑖𝑛𝑖 = 1 − ∑_{i=1}^{n}p_{i}^{2}$

onde:
$p_{i}$ é a proporção de amostras pertencentes à classe $i$;
$n$ é o número total de classes.

O Índice de Gini varia entre **0 e 0.5**:
- **0**: O nó é completamente puro (todas as amostras pertencem à mesma classe).
- **0.5**: O nó é completamente impuro (as amostras são distribuídas uniformemente entre as classes).

---

### **Exemplo Intuitivo**

Imagine um nó com 10 amostras, distribuídas entre duas classes: A e B.

1. **Caso 1: Nó Puro**
   - Todas as 10 amostras pertencem à classe A.
   - $p_A = 1$, $p_B = 0$. Daí, $Gini = 1 - (1^2 + 0^2) = 1 - 1 = 0$. O índice de Gini é 0, indicando um nó puro.

2. **Caso 2: Nó Impuro**
   - 5 amostras pertencem à classe A e 5 à classe B.
   - $p_A = 0.5$, $p_B = 0.5$. Daí, $Gini = 1 - (0.5^2 + 0.5^2) = 1 - (0.25 + 0.25) = 1 - 0.5 = 0.5$. O índice de Gini é 0.5, indicando máxima impureza.

3. **Caso 3: Nó Parcialmente Puro**
   - 7 amostras pertencem à classe A e 3 à classe B.
   - $p_A = 0.7$ e $p_B = 0.3$. Daí, $Gini = 1 - (0.7^2 + 0.3^2) = 1 - (0.49 + 0.09) = 1 - 0.58 = 0.42$. O índice de Gini é 0.42, indicando uma impureza moderada.

---

### **Como o Gini é Usado em Árvores de Decisão**

1. **Divisão Inicial**:
   - O algoritmo calcula o Índice de Gini para cada possível divisão.
   - A divisão que **reduz mais a impureza total** do nó pai para os nós filhos é escolhida.

2. **Cálculo do Gini Ponderado**:
   Se um nó pai é dividido em dois nós filhos, o Gini ponderado é calculado como $Gini_{ponderado} = \frac{N_1}{N} \cdot Gini_1 + \frac{N_2}{N} \cdot Gini_2$
   - $(N_1, N_2)$: Número de amostras nos nós filhos.
   - $(Gini_1, Gini_2)$: Índices de Gini dos nós filhos.
   - $N$: Número total de amostras no nó pai.

3. **Escolha da Divisão**:
   - A divisão com o menor Gini ponderado é selecionada.

---

### **Por Que o Índice de Gini é Importante?**
- Ele garante que as divisões feitas pela árvore de decisão criem nós mais homogêneos.
- É eficiente em termos computacionais, sendo mais rápido de calcular do que outros critérios, como a entropia.
- É amplamente usado em tarefas de classificação, como no exemplo do dataset **Iris**.
