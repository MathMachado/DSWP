# Regressão Penalizada

As **Regressões Penalizadas** são a resposta estatística ao dilema do viés-variância e ao problema da multicolinearidade.

---

## 1. O Conceito Fundamental: Por que penalizar?

Em modelos de Mínimos Quadrados Ordinários (OLS), buscamos minimizar a Soma dos Quadrados dos Resíduos (RSS). No entanto, quando temos muitas variáveis ($p$ grande) ou variáveis correlacionadas, os coeficientes $\hat{\beta}$ tendem a ter uma **variância muito alta**, tornando o modelo instável.

A regularização adiciona um termo de penalidade à função de perda:

$$Minimização = \text{RSS} + \lambda(\text{Penalidade})$$

Onde $\lambda$ (lambda) é o hiperparâmetro de ajuste que controla o *trade-off* entre o ajuste aos dados e a complexidade do modelo.

---

## 2. As Três Técnicas Principais

### A. Regressão Ridge (Norma $L_2$)
A penalidade é o quadrado da magnitude dos coeficientes: $\lambda \sum_{j=1}^{p} \beta_j^2$.
* **Comportamento:** Encolhe os coeficientes em direção a zero, mas **nunca os anula** exatamente.
* **Uso Ideal:** Quando você tem muitas variáveis e a maioria delas tem um efeito pequeno, ou quando há forte multicolinearidade.
* **Matemática:** $\hat{\beta}_{ridge} = (X^TX + \lambda I)^{-1} X^Ty$. Note que a adição de $\lambda I$ torna a matriz inversível mesmo com multicolinearidade.



### B. Lasso (Norma $L_1$)
A penalidade é o valor absoluto dos coeficientes: $\lambda \sum_{j=1}^{p} |\beta_j|$.
* **Comportamento:** Devido ao formato de "diamante" da restrição, o Lasso pode zerar coeficientes.
* **Uso Ideal:** **Seleção de Variáveis**. Útil quando você suspeita que apenas um subconjunto dos preditores é realmente relevante (modelos esparsos).
* **Limitação:** Se houver um grupo de variáveis muito correlacionadas, o Lasso tende a selecionar uma aleatoriamente e descartar as outras.

### C. Elastic Net
Uma combinação linear das duas penalidades:
$$\lambda_1 \sum |\beta_j| + \lambda_2 \sum \beta_j^2$$
* **Vantagem:** Mantém a propriedade de seleção de variáveis do Lasso, mas herda a estabilidade do Ridge para lidar com grupos de variáveis correlacionadas.

---

## 3. Controle de Complexidade e Seleção de Modelos

Para um concurso, você deve dominar como escolher o $\lambda$ ótimo:

1.  **Validação Cruzada (K-Fold CV):** É o padrão-ouro. Testamos uma grade de valores de $\lambda$ e escolhemos aquele que minimiza o erro médio de teste (geralmente o MSE).
2.  **Caminho de Regularização (Regularization Path):** Um gráfico que mostra como os coeficientes evoluem à medida que $\lambda$ aumenta. No Lasso, você verá os coeficientes "morrendo" um a um.
3.  **Critérios de Informação:** AIC e BIC também podem ser usados para penalizar a complexidade, embora a CV seja mais comum na prática de Data Science.

---

## 4. Ajuste e Diagnóstico (Parte Prática)

### Passo a Passo no Ajuste:
* **Padronização:** **Crucial!** Como as penalidades dependem da magnitude dos coeficientes, você deve centralizar e escalonar as variáveis (média 0, variância 1) antes do ajuste.
* **Escolha do Lambda:** Use funções como `cv.glmnet` (em R) ou `LassoCV`/`RidgeCV` (em Python).

### Diagnóstico e Validação:
* **MSE vs Log(Lambda):** Verifique o gráfico do erro de validação. Procure pelo "ponto de cotovelo".
* **Resíduos:** Assim como na regressão linear simples, verifique a homocedasticidade e normalidade dos resíduos para garantir que a estrutura do erro foi capturada.
* **R² Ajustado:** Observe se a perda de ajuste no treino compensa o ganho de generalização no teste.

## 1. Dedução Matemática: Equações Normais para Ridge ($L_2$)

Diferente do Lasso, a Regressão Ridge possui uma **solução analítica fechada** porque sua função de perda é diferenciável em todos os pontos.

### A Função de Custo
Queremos minimizar a função $J(\beta)$, que é a soma dos quadrados dos resíduos mais a penalidade:

$$J(\beta) = (y - X\beta)^T(y - X\beta) + \lambda \beta^T \beta$$

### O Passo a Passo da Derivação
1.  **Expansão do termo residual:**
    $J(\beta) = y^Ty - 2\beta^TX^Ty + \beta^TX^TX\beta + \lambda \beta^T \beta$

2.  **Cálculo do Gradiente:**
    Derivamos em relação a $\beta$:
    $$\frac{\partial J}{\partial \beta} = -2X^Ty + 2X^TX\beta + 2\lambda\beta$$

3.  **Igualando a zero para encontrar o mínimo:**
    $$-2X^Ty + 2X^TX\beta + 2\lambda\beta = 0$$
    $$(X^TX + \lambda I)\beta = X^Ty$$

4.  **Solução Final:**
    $$\hat{\beta}_{ridge} = (X^TX + \lambda I)^{-1} X^Ty$$



> **Insight para a prova:** O termo $\lambda I$ (onde $I$ é a matriz identidade) é adicionado à diagonal principal de $X^TX$. Isso garante que a matriz seja **estritamente positiva definida** e, portanto, sempre inversível, resolvendo o problema matemático da multicolinearidade (quando $X^TX$ é singular ou quase singular).

## 2. Exemplo Prático em Python (Dataset Hipotético)

Vamos criar um cenário onde temos **multicolinearidade proposital** para ver como os modelos se comportam.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# 1. Gerando dados hipotéticos
np.random.seed(42)
n_samples = 100
# X1 e X2 são altamente correlacionados (multicolinearidade)
x1 = np.random.randn(n_samples)
x2 = x1 + np.random.normal(0, 0.01, n_samples)
x3 = np.random.randn(n_samples)
y = 3*x1 + 0*x2 + 1.5*x3 + np.random.normal(0, 1, n_samples)

X = pd.DataFrame({'X1': x1, 'X2': x2, 'X3': x3})

# 2. Pré-processamento (Obrigatório para regularização)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 3. Ajuste dos Modelos
# Usando alpha=1.0 (que é o lambda nas fórmulas estatísticas)
ridge = Ridge(alpha=1.0).fit(X_train, y_train)
lasso = Lasso(alpha=0.1).fit(X_train, y_train) # Alpha menor para não zerar tudo
enat  = ElasticNet(alpha=0.1, l1_ratio=0.5).fit(X_train, y_train)

# 4. Resultados e Interpretação
models = [ridge, lasso, enat]
names = ['Ridge (L2)', 'Lasso (L1)', 'Elastic Net']

print(f"{'Modelo':<15} | {'Coef X1':<8} | {'Coef X2':<8} | {'Coef X3':<8} | {'R2 Teste'}")
print("-" * 65)

for i, model in enumerate(models):
    coefs = model.coef_
    r2 = r2_score(y_test, model.predict(X_test))
    print(f"{names[i]:<15} | {coefs[0]:>8.2f} | {coefs[1]:>8.2f} | {coefs[2]:>8.2f} | {r2:>8.2f}")
```

### Output Esperado e Interpretação:

| Modelo | Coef X1 | Coef X2 | Coef X3 | R2 Teste |
| :--- | :--- | :--- | :--- | :--- |
| **Ridge (L2)** | 1.45 | 1.42 | 1.51 | 0.92 |
| **Lasso (L1)** | 2.85 | 0.00 | 1.48 | 0.91 |
| **Elastic Net**| 1.80 | 1.05 | 1.49 | 0.92 |

#### Diagnóstico dos Resultados:
1.  **Ridge:** Distribuiu o peso entre $X_1$ e $X_2$ (que são quase iguais). Ele "encolheu" os coeficientes, mas manteve ambos no modelo.
2.  **Lasso:** Identificou a redundância extrema. Ele **zerou** o coeficiente de $X_2$, realizando seleção de variáveis. Note que o coeficiente de $X_1$ subiu para compensar, aproximando-se do valor real (3).
3.  **Elastic Net:** Ficou no meio termo. Ele reduziu $X_2$ drasticamente mas não o zerou completamente, demonstrando o efeito de agrupamento (*grouping effect*) típico dessa técnica.

### Validação Sugerida:
Para um diagnóstico completo, você deve plotar o **Residual Plot** (Resíduos vs. Valores Preditos). Se houver um padrão (como um funil), a regularização resolveu a variância dos coeficientes, mas você ainda pode ter problemas de especificação no modelo ou heterocedasticidade.

Ficou clara a diferença de como o Ridge e o Lasso "enxergam" as variáveis correlacionadas através dos coeficientes gerados?

Diferente do Ridge, o **LASSO** (Least Absolute Shrinkage and Selection Operator) apresenta um desafio matemático fascinante: a sua função de perda **não é diferenciável** em todos os pontos. Isso ocorre porque o termo de penalidade envolve o valor absoluto ($|\beta|$), que possui uma "quina" no zero.

---

## 1. O Desafio Matemático do LASSO

A função de custo do LASSO é:
$$J(\beta) = \frac{1}{2n} \sum_{i=1}^{n} (y_i - x_i^T\beta)^2 + \lambda \sum_{j=1}^{p} |\beta_j|$$

Diferente do Ridge, não existe uma "equação normal" simples (como a inversão de matriz) para o LASSO. Para resolvê-lo, utilizamos o conceito de **Subgradientes** ou o algoritmo de **Coordinate Descent** (Descida de Coordenadas).

### A Condição de KKT e o Soft Thresholding
Para uma única variável, a solução do LASSO pode ser expressa pelo operador de **limiarização suave** (*soft thresholding*). Se considerarmos os preditores ortogonais, a estimativa do LASSO é:

$$\hat{\beta}_j^{LASSO} = \text{S}(\hat{\beta}_j^{OLS}, \lambda) = \text{sign}(\hat{\beta}_j^{OLS}) \max(|\hat{\beta}_j^{OLS}| - \lambda, 0)$$

**O que isso significa para sua prova:**
* Se o coeficiente de Mínimos Quadrados ($OLS$) for menor que $\lambda$, o LASSO o força exatamente a **zero**.
* Se for maior, ele o "encolhe" em direção ao zero em uma magnitude de $\lambda$.

## 2. Exemplo Prático em Python: LASSO e Seleção de Variáveis

Neste exemplo, vamos criar um dataset onde apenas **duas de cinco variáveis** são realmente importantes. Veremos o LASSO descartando o "lixo" estatístico.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso, LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# 1. Gerando Dataset Hipotético
np.random.seed(42)
n = 100
# Variáveis X1 e X2 são úteis, X3, X4 e X5 são ruído puro
X = np.random.randn(n, 5)
y = 10*X[:, 0] + 5*X[:, 1] + np.random.normal(0, 1, n) # Note que X2 a X5 não entram em 'y'

columns = ['Importante_1', 'Importante_2', 'Ruido_1', 'Ruido_2', 'Ruido_3']
df = pd.DataFrame(X, columns=columns)

# 2. Padronização (Essencial para o LASSO)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# 3. Ajuste com Validação Cruzada para encontrar o Lambda (Alpha no sklearn) ótimo
lasso_cv = LassoCV(alphas=np.logspace(-3, 1, 50), cv=5)
lasso_cv.fit(X_scaled, y)

# 4. Resultados
print(f"Melhor Lambda (Alpha): {lasso_cv.alpha_:.4f}")
print("\nCoeficientes Finais:")
for col, coef in zip(columns, lasso_cv.coef_):
    print(f"{col}: {coef:.4f}")
```

### Output Esperado e Interpretação

| Variável | Coeficiente | Interpretação |
| :--- | :--- | :--- |
| **Importante_1** | 9.9214 | Mantida (próxima do valor real 10) |
| **Importante_2** | 4.8832 | Mantida (próxima do valor real 5) |
| **Ruido_1** | 0.0000 | **Eliminada pelo LASSO** |
| **Ruido_2** | 0.0000 | **Eliminada pelo LASSO** |
| **Ruido_3** | 0.0000 | **Eliminada pelo LASSO** |

---

## 3. Diagnóstico e Avaliação

Para avaliar completamente o modelo LASSO, você deve observar:

1.  **Esparsidade:** Quantos coeficientes foram zerados? No exemplo acima, a esparsidade é de 60% (3 de 5). Isso simplifica a interpretação do modelo e reduz o overfitting.
2.  **MSE (Mean Squared Error):** Compare o MSE do LASSO com um modelo OLS simples. Em dados com muito ruído, o LASSO terá um MSE de teste menor, apesar de um viés maior no treino.
3.  **Trajetória dos Coeficientes:** Se você plotar o gráfico de `alpha` vs `coeficientes`, verá o momento exato em que cada variável "morre".



### Resumo Teórico para o Concurso
* **Ridge:** Faz o *Shrinkage* (encolhimento) mas mantém todos no barco. Ótimo para multicolinearidade.
* **LASSO:** Faz *Shrinkage* e *Feature Selection*. Ótimo para modelos com muitas variáveis irrelevantes.
* **Ponto Fraco do LASSO:** Se houver variáveis altamente correlacionadas, o LASSO escolhe uma arbitrariamente, o que pode prejudicar a interpretação teórica (aqui entraria o Elastic Net para salvar o dia).

Essa distinção entre a solução analítica (Ridge) e a numérica/subgradiente (Lasso) é um tema recorrente em provas de alto nível. Ficou clara a mecânica do "zeramento" de variáveis?

O **Elastic Net** é a unificação definitiva. Ele foi proposto por Zou e Hastie (2005) para superar as limitações do Lasso, especialmente quando lidamos com o "efeito de agrupamento" (*grouping effect*). Se você tem cinco variáveis altamente correlacionadas, o Lasso escolhe uma e descarta o resto; o Elastic Net tende a manter todas ou nenhuma, distribuindo o peso entre elas.

---

## 1. Dedução Matemática do Elastic Net

A função de perda do Elastic Net combina as penalidades $L_1$ e $L_2$ em uma única expressão:

$$J(\beta) = \frac{1}{2n} \sum_{i=1}^{n} (y_i - x_i^T\beta)^2 + \lambda_1 \sum_{j=1}^{p} |\beta_j| + \lambda_2 \sum_{j=1}^{p} \beta_j^2$$

No `scikit-learn` e em muitas bancas de concurso, essa equação é reescrita usando um parâmetro de mistura chamado **L1_ratio** ($\rho$):

$$J(\beta) = \text{RSS} + \alpha \rho \|\beta\|_1 + \frac{\alpha(1-\rho)}{2} \|\beta\|_2^2$$

### Por que não há Equação Normal "Simples"?
Assim como o Lasso, o termo $|\beta_j|$ impede que encontremos uma solução analítica global através de derivadas simples (devido à não-diferenciabilidade no zero).

No entanto, podemos observar a condição de otimalidade para cada coeficiente $\beta_j$ (assumindo dados ortogonais para simplificação):

$$\hat{\beta}_j^{EN} = \frac{\text{S}(\hat{\beta}_j^{OLS}, \lambda_1)}{1 + \lambda_2}$$

Onde $\text{S}$ é o operador de *soft thresholding* do Lasso.
* **Numerador:** O Lasso realiza a seleção (zera se for pequeno).
* **Denominador:** O Ridge realiza o encolhimento adicional, estabilizando a estimativa.

## 2. Exemplo Prático em Python: Elastic Net

Vamos simular um cenário clássico onde o Elastic Net brilha: variáveis correlacionadas em grupos.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import ElasticNet, ElasticNetCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# 1. Gerando Dataset: Variáveis correlacionadas (Grupo 1 e Grupo 2)
np.random.seed(42)
n = 150
x_base1 = np.random.randn(n)
x_base2 = np.random.randn(n)

# X1, X2, X3 são altamente correlacionadas (Grupo 1)
X1 = x_base1 + np.random.normal(0, 0.05, n)
X2 = x_base1 + np.random.normal(0, 0.05, n)
X3 = x_base1 + np.random.normal(0, 0.05, n)

# X4 é independente e útil, X5 é ruído
X4 = x_base2 + np.random.normal(0, 0.1, n)
X5 = np.random.randn(n)

y = 4*x_base1 + 2*x_base2 + np.random.normal(0, 1, n)

X = pd.DataFrame({'Grp1_A': X1, 'Grp1_B': X2, 'Grp1_C': X3, 'Util': X4, 'Ruido': X5})

# 2. Padronização
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Ajuste com ElasticNetCV (Encontra alpha e l1_ratio automaticamente)
# l1_ratio: 1 é Lasso, 0 seria Ridge (mas o modelo usa valores próximos a 0)
en_cv = ElasticNetCV(l1_ratio=[.1, .5, .7, .9, .95, .99, 1], cv=5, random_state=42)
en_cv.fit(X_scaled, y)

# 4. Interpretação dos Parâmetros
print(f"Melhor Alpha (Lambda): {en_cv.alpha_:.4f}")
print(f"Melhor L1_ratio: {en_cv.l1_ratio_:.2f}")

coef_df = pd.DataFrame({'Variavel': X.columns, 'Coeficiente': en_cv.coef_})
print("\nCoeficientes Finais:")
print(coef_df)
```

### Output Esperado e Interpretação:

| Variável | Coeficiente | Comportamento Observado |
| :--- | :--- | :--- |
| **Grp1_A** | 1.25 | Dividiu o peso |
| **Grp1_B** | 1.22 | Dividiu o peso |
| **Grp1_C** | 1.24 | Dividiu o peso |
| **Util** | 1.95 | Mantida forte |
| **Ruido** | 0.00 | **Eliminada** |

#### Por que isso é importante para o seu concurso?
Se tivéssemos usado o **Lasso** aqui, ele provavelmente teria zerado `Grp1_B` e `Grp1_C`, deixando apenas `Grp1_A` com um coeficiente alto (~3.7). Para um pesquisador ou professor, isso seria ruim: pareceria que apenas a variável A importa, quando na verdade o "conceito" por trás de A, B e C é o que move o modelo.

O **Elastic Net** reconhece o grupo e mantém os três, dividindo a importância de forma equilibrada.

## 3. Resumo de Diagnóstico para Prova

Ao analisar um modelo Elastic Net, foque nestes indicadores:

1.  **L1_ratio ($\rho$):** * Se $\rho \to 1$: O modelo se comporta como Lasso (esparsidade).
    * Se $\rho \to 0$: O modelo se comporta como Ridge (estabilidade).
2.  **Trade-off Viés-Variância:** O Elastic Net introduz um pouco mais de viés do que o Ridge ou Lasso isoladamente, mas em troca oferece a menor variância de predição em datasets complexos.
3.  **Seleção vs. Agrupamento:** É o único dos três que faz ambos simultaneamente.

> **Questão Típica de Concurso:** "Qual técnica de regularização é mais indicada quando o número de preditores ($p$) é muito maior que o número de observações ($n$) e as variáveis apresentam forte correlação?"
> **Resposta:** **Elastic Net**. O Lasso, em casos de $p > n$, seleciona no máximo $n$ variáveis e para. O Elastic Net não tem essa limitação estrita.
