# Guia de Revisão: Support Vector Machines (SVM)

## 1. O Conceito Fundamental: Hiperplanos de Separação
O objetivo básico do SVM é encontrar um **hiperplano de separação ótimo** entre duas classes[cite: 533].
* **Margem:** É a distância entre o hiperplano de decisão e os pontos de treinamento mais próximos de cada classe[cite: 569]. O SVM busca **maximizar esta margem** ($M$) para garantir uma melhor generalização[cite: 563, 565].
* **Vetor de Pesos ($\beta$):** O hiperplano é definido por $\{x : f(x) = x^T\beta + \beta_0 = 0\}$[cite: 558]. A largura da margem é dada por $2M = 2/||\beta||$[cite: 554, 574].

## 2. O Support Vector Classifier (Caso Linear)
Quando as classes não são perfeitamente separáveis (há sobreposição), o SVM utiliza o conceito de **margem suave (soft margin)**[cite: 577].

* **Variáveis de Folga (Slack Variables, $\xi_i$):** Permitem que alguns pontos fiquem do lado errado da margem ou até do lado errado do hiperplano[cite: 578, 587].
    * $\xi_i = 0$: Ponto classificado corretamente fora da margem[cite: 941].
    * $0 < \xi_i \le 1$: Ponto dentro da margem, mas do lado correto da decisão[cite: 942, 944].
    * $\xi_i > 1$: Ponto mal classificado (lado errado da decisão)[cite: 589].
* **Custo (Parâmetro $C$):** É o parâmetro de tunagem que controla o balanço entre maximizar a margem e minimizar o erro de treinamento[cite: 606, 641].
    * $C$ grande: Penaliza severamente as folgas, focando em pontos próximos à fronteira (pode causar *overfitting*)[cite: 716].
    * $C$ pequeno: Permite uma margem mais larga e suave, ignorando erros individuais em favor de uma fronteira mais simples[cite: 717].


## 3. Support Vector Machines e Kernels
Para lidar com fronteiras de decisão não lineares, o SVM projeta os dados em um espaço de características de alta dimensão usando funções de base[cite: 535, 677].

* **O Truque do Kernel (Kernel Trick):** O SVM pode ser resolvido utilizando apenas o **produto interno** entre os vetores no novo espaço, sem precisar calcular explicitamente a transformação[cite: 687, 698]. O kernel é a função $K(x, x') = \langle h(x), h(x') \rangle$[cite: 699].
* **Kernels Populares:**
    1.  **Polinomial:** $K(x, x') = (1 + \langle x, x' \rangle)^d$[cite: 703].
    2.  **Radial Basis (RBF/Gaussiano):** $K(x, x') = \exp(-\gamma ||x - x'||^2)$[cite: 704].
    3.  **Sigmoide (Neural Network):** $K(x, x') = \tanh(\kappa_1 \langle x, x' \rangle + \kappa_2)$[cite: 704].

## 4. Por que "Vetores de Suporte"?
A solução final do SVM depende apenas de um subconjunto dos pontos de treinamento: os **vetores de suporte**[cite: 635].
* São os pontos que estão exatamente na margem ($\xi_i = 0$) ou que violam a margem ($\xi_i > 0$)[cite: 636].
* Pontos que estão corretamente classificados e fora da margem não influenciam a posição do hiperplano[cite: 652, 941]. Isso torna o SVM robusto a *outliers* que estejam longe da fronteira de decisão[cite: 595].

## 5. Formulação Matemática (Dualidade)
O problema de otimização é um problema quadrático convexo[cite: 601].
* **Função de Perda Hinge (Hinge Loss):** O SVM minimiza a perda dada por $[1 - y_if(x_i)]_+$, onde o subscrito "+" indica a parte positiva[cite: 761, 764]. Esta função dá penalidade zero para pontos fora da margem e uma penalidade linear para pontos que a violam[cite: 766, 773].


## 6. SVM para Regressão (SVR)
O SVM também pode ser adaptado para respostas quantitativas[cite: 957].
* **$\epsilon$-Insensitive Error:** O modelo ignora erros menores que um valor $\epsilon$, criando um "tubo" ao redor da função de regressão[cite: 984, 986]. Apenas os pontos fora deste tubo (com resíduos grandes) atuam como vetores de suporte[cite: 1006].

---

### Dicas para a Prova:
1.  **LDA vs. SVM:** No LDA, a fronteira depende das médias e covariâncias de todos os pontos de todas as classes[cite: 597]. No SVM, a fronteira depende apenas dos pontos próximos a ela (vetores de suporte)[cite: 595].
2.  **Custo $C$:** Lembre-se que $C$ substitui o multiplicador de Lagrange. Se a questão falar em penalização $(\lambda)$, a relação é inversa: $C = 1/\lambda$[cite: 763].
3.  **Kernel RBF:** É muito sensível ao parâmetro $\gamma$ e ao custo $C$. Use validação cruzada para encontrá-los[cite: 896, 931].

# SVM

O **SVM (Support Vector Machines)** como uma das construções mais bonitas da área. Ele une otimização convexa, geometria e o "truque" matemático mais famoso da computação: o Kernel.

Vamos por partes, começando do fundamento linear até a complexidade do aprendizado estatístico.

---

## 1. Teoria Estatística do Aprendizado (SLT)

O SVM não surgiu por tentativa e erro; ele é fundamentado na **Teoria de Vapnik-Chervonenkis (VC)**. O objetivo aqui não é apenas minimizar o erro no treino (Risco Empírico), mas minimizar o **Risco Estrutural**.

A ideia central é que a capacidade de generalização de um modelo depende de:
1.  **Erro de treinamento.**
2.  **Dimensão VC** (uma medida da complexidade/flexibilidade do modelo).

O SVM busca o equilíbrio ao maximizar a **margem**, o que tecnicamente limita a Dimensão VC, garantindo que o modelo não "decore" o ruído.

---

## 2. O Classificador de Margem Máxima (Caso Linearmente Separável)

Imagine dois conjuntos de pontos que podem ser divididos por uma reta (ou hiperplano). Existem infinitas retas possíveis, mas o SVM escolhe aquela que fica o mais longe possível dos pontos mais próximos de cada classe.

### A Dedução Matemática

Seja o hiperplano definido por:
$$w^T x + b = 0$$

Queremos que para cada ponto $x_i$ com rótulo $y_i \in \{-1, 1\}$:
* Se $y_i = 1$, então $w^T x_i + b \ge 1$
* Se $y_i = -1$, então $w^T x_i + b \le -1$

Isso pode ser resumido na restrição:
$$y_i(w^T x_i + b) \ge 1, \quad \forall i$$



**A Margem:**
A distância de um ponto ao hiperplano é dada por $\frac{|w^T x + b|}{\|w\|}$. Para os pontos onde a restrição é igual a 1 (os vetores de suporte), a distância é $1/\|w\|$. A margem total é $2/\|w\|$.

Para **maximizar a margem**, precisamos **minimizar $\|w\|$**. Por conveniência matemática, minimizamos o quadrado:

> **Problema Primal (Hard Margin):**
> $$\min_{w, b} \frac{1}{2} \|w\|^2 \quad \text{sujeito a} \quad y_i(w^T x_i + b) \ge 1$$

---

## 3. Soft Margin: Lidando com o Ruído

Na vida real, os dados se sobrepõem. Para evitar o overfitting e permitir erros controlados, introduzimos as **variáveis de folga (slack variables)** $\xi_i \ge 0$.

A nova restrição torna-se:
$$y_i(w^T x_i + b) \ge 1 - \xi_i$$

Onde $\xi_i$ mede o quanto o ponto invade a margem. O novo objetivo é:
$$\min_{w, b, \xi} \frac{1}{2} \|w\|^2 + C \sum_{i=1}^n \xi_i$$

O parâmetro **$C$** (Tuning) controla o tradeoff:
* **$C$ alto:** Penaliza muito o erro. O modelo tenta classificar tudo corretamente (Margem Estreita, risco de Overfitting).
* **$C$ baixo:** Tolera mais erros em troca de uma margem mais larga (Maior Generalização, risco de Underfitting).

---

## 4. O Dual de Lagrange e a Função Kernel

Para resolver o problema acima e introduzir não-linearidade, usamos multiplicadores de Lagrange $\alpha_i$. O **Lagrangeano** é:
$$L(w, b, \xi, \alpha, \mu) = \frac{1}{2}\|w\|^2 + C\sum \xi_i - \sum \alpha_i[y_i(w^Tx_i + b) - 1 + \xi_i] - \sum \mu_i \xi_i$$

Ao derivar em relação a $w$ e $b$ e igualar a zero, chegamos ao **Problema Dual**:

> **Dual de Lagrange:**
> $$\max_{\alpha} \sum_{i=1}^n \alpha_i - \frac{1}{2} \sum_{i,j=1}^n \alpha_i \alpha_j y_i y_j (x_i^T x_j)$$
> Sujeito a: $0 \le \alpha_i \le C$ e $\sum \alpha_i y_i = 0$.

### O Truque do Kernel (The Kernel Trick)
Note que os dados $x_i$ só aparecem na fórmula como um **produto escalar** $(x_i^T x_j)$. Se quisermos mapear os dados para um espaço de alta dimensão $\phi(x)$ para torná-los separáveis, não precisamos calcular $\phi(x)$ explicitamente. Basta substituir o produto escalar por uma **Função Kernel**:
$$K(x_i, x_j) = \phi(x_i)^T \phi(x_j)$$

**Kernels comuns:**
* **Linear:** $K(x, z) = x^T z$
* **Polinomial:** $(x^T z + r)^d$
* **RBF (Gaussiano):** $\exp(-\gamma \|x - z\|^2)$ — O parâmetro $\gamma$ controla o alcance da influência de um único ponto.



---

## 5. Prática: Tuning e Validação

Para concursos, lembre-se do processo de **Grid Search** com **Cross-Validation**:
1.  **Escalonamento:** SVM é sensível à escala (usa distâncias euclidianas). Sempre padronize os dados ($z$-score).
2.  **Busca de Hiperparâmetros:** No RBF, ajustamos $C$ e $\gamma$ simultaneamente.
3.  **Validação:** K-fold Cross-Validation é o padrão ouro para estimar o erro de generalização.

## 6. Aplicações em Grandes Volumes e Limitações

### Grandes Volumes de Dados
O SVM original tem complexidade aproximada de $O(n^2)$ a $O(n^3)$, o que o torna proibitivo para milhões de linhas. Soluções:
* **Linear SVM (LibLinear):** Otimizado para kernel linear em dados esparsos (ex: classificação de texto).
* **Stochastic Gradient Descent (SGD):** Aproxima a solução do SVM de forma incremental.

### Limitações
1.  **Não probabilístico:** O SVM não fornece probabilidades diretamente (usa-se o método de Platt para aproximar).
2.  **Sensibilidade a ruído:** Embora o Soft Margin ajude, labels trocados podem distorcer os vetores de suporte.
3.  **Custo Computacional:** Lento para conjuntos de dados massivos com kernels não-lineares.
4.  **Interpretabilidade:** No espaço do Kernel, é difícil explicar "por que" uma decisão foi tomada.

---

### Dica de Especialista para a Prova:
Muitas questões de concurso focam nas **Condições KKT (Karush-Kuhn-Tucker)**. Lembre-se que apenas os pontos onde $\alpha_i > 0$ são os **Vetores de Suporte**. Eles são os únicos pontos que realmente "sustentam" a solução; se você remover todos os outros pontos do dataset e treinar novamente, o resultado será idêntico.

Vamos deduzir o fechamento do problema para o caso de **Margem Rígida (Hard Margin)**, pois a lógica para a Margem Suave é análoga, apenas adicionando restrições aos multiplicadores.

## 1. Dedução Matemática: Do Primal ao Dual

Nosso objetivo é minimizar a norma do vetor de pesos, sujeita à classificação correta de todos os pontos:

$$\min_{w, b} \frac{1}{2} \|w\|^2 \quad \text{s.t.} \quad g_i(w, b) = 1 - y_i(w^T x_i + b) \le 0$$

### Passo 1: Construção do Lagrangeano
Introduzimos os multiplicadores de Lagrange $\alpha_i \ge 0$ para cada restrição:

$$L(w, b, \alpha) = \frac{1}{2} w^T w + \sum_{i=1}^n \alpha_i [1 - y_i(w^T x_i + b)]$$

### Passo 2: Condições de Primeira Ordem (KKT)
Para encontrar o mínimo em relação a $w$ e $b$, derivamos e igualamos a zero:

1.  **Derivada em relação a $w$:**
    $$\nabla_w L = w - \sum_{i=1}^n \alpha_i y_i x_i = 0 \implies \mathbf{w = \sum_{i=1}^n \alpha_i y_i x_i}$$
    *Insight:* O vetor de pesos é uma combinação linear dos vetores de treino.

2.  **Derivada em relação a $b$:**
    $$\frac{\partial L}{\partial b} = - \sum_{i=1}^n \alpha_i y_i = 0 \implies \mathbf{\sum_{i=1}^n \alpha_i y_i = 0}$$

### Passo 3: Substituição para a Forma Dual
Agora, substituímos $w$ de volta na equação de $L(w, b, \alpha)$ para eliminar as variáveis primais.

$$L = \frac{1}{2} \left( \sum \alpha_i y_i x_i \right)^T \left( \sum \alpha_j y_j x_j \right) + \sum \alpha_i - \sum \alpha_i y_i \left( \sum \alpha_j y_j x_j \right)^T x_i - b \underbrace{\sum \alpha_i y_i}_{0}$$

Após simplificar os termos (o primeiro termo é $1/2 \sum \sum \dots$ e o terceiro é $-\sum \sum \dots$), chegamos à **Função Dual de Lagrange**:

$$W(\alpha) = \sum_{i=1}^n \alpha_i - \frac{1}{2} \sum_{i=1}^n \sum_{j=1}^n \alpha_i \alpha_j y_i y_j (x_i^T x_j)$$

**O que isso significa?**
O problema agora depende apenas de $\alpha$. Os pontos $x_i$ e $x_j$ aparecem apenas como um **produto escalar**. É aqui que entra o **Kernel Trick**: se os dados não forem linearmente separáveis, trocamos $(x_i^T x_j)$ por $K(x_i, x_j)$.

## 2. Implementação Prática em Python

Para concursos, a biblioteca padrão é a `scikit-learn`. Vamos implementar um exemplo clássico com o Kernel RBF (Radial Basis Function) e visualização de margens.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs

# 1. Gerar dados sintéticos (quase separáveis)
X, y = make_blobs(n_samples=50, centers=2, random_state=6)

# 2. Ajustar o modelo SVM
# C=1.0 (Soft Margin), kernel='rbf' (Não linear), gamma='scale'
clf = svm.SVC(kernel='rbf', C=1.0, gamma=0.7)
clf.fit(X, y)

# 3. Visualização básica
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# Plotar os Vetores de Suporte (aqueles onde alpha > 0)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
            linewidth=1, facecolors='none', edgecolors='k', label='Vetores de Suporte')

# Criar grade para desenhar a fronteira de decisão
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 30),
                     np.linspace(ylim[0], ylim[1], 30))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Desenhar margens e hiperplano
ax.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

plt.title("SVM com Kernel RBF e Margens Suaves")
plt.legend()
plt.show()
```

### Análise do Código para a Prova:
* `clf.support_vectors_`: Retorna os pontos que definem a margem. Em grandes datasets, se o número de vetores de suporte for muito alto, o modelo pode estar sofrendo **overfitting**.
* `decision_function`: Calcula a distância com sinal de cada amostra até o hiperplano. O sinal indica a classe e a magnitude indica a confiança.

## 3. Resumo para Revisão Rápida

| Conceito | O que levar para a prova |
| :--- | :--- |
| **Primal** | Minimiza $\|w\|^2$ (Geometria do hiperplano). |
| **Dual** | Maximiza em relação a $\alpha$ (Depende apenas do produto escalar entre pontos). |
| **Kernel** | Permite projetar dados em dimensões infinitas (RBF) sem computar as coordenadas. |
| **Parâmetro C** | Inversamente proporcional à regularização. C grande = Margem rígida. |
| **Vetor de Suporte** | Pontos onde $y_i(w^T x_i + b) = 1$ (ou $1-\xi_i$). São os únicos que importam para o modelo final. |

## 2. Interpretando um Caso Prático

A função de decisão do SVM é dada por $f(x) = \text{sign}(w^T x + b)$. Vamos testar para um funcionário específico (com dados já padronizados):

> **Perfil do Desenvolvedor "João":**
> * Satisfação: $-1$ (Um desvio padrão abaixo da média)
> * Projetos: $+2$ (Muito acima da média)
> * Anos_Empresa: $0$ (Na média)
> * Salario_Mensal: $-0.5$ (Pouco abaixo da média)
> * Horas_Extras: $+2$ (Muitas horas extras)

### O Cálculo:
$$f(x) = [(-1.25 \times -1) + (0.45 \times 2) + (-0.10 \times 0) + (-0.85 \times -0.5) + (1.50 \times 2)] - 0.30$$
$$f(x) = [1.25 + 0.90 + 0 + 0.425 + 3.0] - 0.30 = \mathbf{5.275}$$

**Resultado:** Como $f(x) > 0$, o modelo classifica que o João **tem alta probabilidade de sair da empresa**.

### Por que isso cai em prova?
1.  **Magnitude dos Pesos:** No SVM linear, o valor absoluto de $w_i$ indica a importância da variável (Feature Importance). Variáveis com $w_i$ próximo de zero quase não afetam a posição da margem.
2.  **Sinal dos Pesos:** O sinal indica a direção. Positivo favorece a classe $1$, negativo favorece a classe $-1$.
3.  **Vetores de Suporte:** Se o João estivesse com $f(x) = 1.0$, ele seria um **Vetor de Suporte**. Ele estaria exatamente no limite da "zona de incerteza" da empresa.
