## 1. Árvores de Decisão (CART)
As árvores de decisão (Classification and Regression Trees) utilizam uma abordagem de particionamento binário recursivo do espaço de características[cite: 1207].

* **Fundamento:** O algoritmo busca subdivisões nas variáveis explicativas ($X$) que maximizem a homogeneidade (pureza) da variável resposta ($Y$) em cada nó resultante[cite: 1207].
* **Critérios de Impureza:** Para classificação, utilizam-se comumente a **Impureza de Gini** ou a **Entropia**. Para regressão, utiliza-se a **Soma dos Quadrados dos Resíduos (SQR)**, de forma análoga à minimização de erros em modelos lineares[cite: 209, 1029].
* **Vantagem Teórica:** Diferente do SVM ou MLGs, as árvores capturam interações de alta ordem e não-linearidades de forma nativa e intuitiva[cite: 537, 1201].



---

## 2. Fundamentos de Ensembles: Bagging
O *Bootstrap Aggregating* (Bagging) é a base para o Random Forest e visa reduzir a **variância** do modelo[cite: 1082].

* **Bootstrap:** Gera-se múltiplos subconjuntos de dados através de amostragem com reposição do conjunto original[cite: 1340, 1400].
* **Agregação:** Treina-se uma árvore independente em cada subconjunto. A predição final é a média (regressão) ou voto majoritário (classificação) das árvores[cite: 1053, 1312].
* **OOB (Out-of-Bag):** Aproximadamente 1/3 dos dados não são usados em cada amostra bootstrap, servindo como um conjunto de teste natural para estimativa de erro sem necessidade de validação cruzada externa inicial[cite: 651].

---

## 3. Random Forest (Florestas Aleatórias)
O Random Forest aprimora o Bagging ao introduzir **aleatoriedade nas variáveis**[cite: 851, 853].

* **Descorrelação:** Em cada divisão de cada árvore, apenas um subconjunto aleatório de $m$ preditores é considerado. Isso evita que preditores muito fortes dominem todas as árvores, reduzindo a correlação entre elas e, consequentemente, a variância do conjunto[cite: 857, 1270].
* **Propriedade de Importância:** Permite ranquear as variáveis com base no quanto elas reduzem a impureza global, similar à análise de significância de preditores em MLGs[cite: 2783, 3061].



---

## 4. Tuning e Validação
A performance de modelos baseados em árvores depende criticamente do ajuste de hiperparâmetros (tuning)[cite: 641, 715, 931].

* **Hiperparâmetros Chave:**
    * `nodesize`: Tamanho mínimo das folhas (controla o viés/variância)[cite: 488].
    * `mtry`: Número de variáveis sorteadas em cada nó (essencial para Random Forest).
    * `ntree`: Número de árvores na floresta.
* **Validação:** Utiliza-se **Validação Cruzada ($k$-fold)** para encontrar os parâmetros ótimos, similar ao processo usado para definir o parâmetro de custo $C$ no SVM ou $\lambda$ no GAM[cite: 475, 651, 931, 1013].

## 5. Aplicações em Grandes Volumes de Dados
Modelos de árvores escalam bem, mas exigem estratégias específicas para *Big Data*[cite: 1476]:

* **Eficiência Computacional:** Enquanto algoritmos como o SVM podem ter complexidade $O(N^3)$, árvores individuais são rápidas ($O(N \log N)$)[cite: 1476, 1480].
* **Paralelização:** Como as árvores no Bagging/Random Forest são independentes, elas podem ser treinadas em paralelo, otimizando o uso de clusters[cite: 1481].
* **Subamostragem:** Em grandes volumes, pode-se usar apenas uma fração dos dados para cada árvore sem perda significativa de acurácia.

## 6. Limitações
Apesar de robustos, esses modelos possuem gargalos teóricos[cite: 1084]:

* **Interpretabilidade:** Enquanto o MLG oferece coeficientes interpretáveis ($\beta$) e as árvores simples são visuais, o Random Forest é um modelo "caixa-preta"[cite: 1080, 2254, 2267].
* **Extrapolação:** Modelos baseados em árvores não conseguem extrapolar tendências fora do intervalo dos dados de treinamento; eles apenas replicam os valores das folhas existentes.
* **Natureza Descontínua:** Diferente da Regressão Logística ou Probit, que geram probabilidades suaves e contínuas, as árvores geram predições em "degraus", o que pode ser inadequado para fenômenos estritamente suaves[cite: 1871, 2275, 2515].
