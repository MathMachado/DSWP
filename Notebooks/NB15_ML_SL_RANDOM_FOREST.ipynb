# **RANDOM FOREST**
* Decision Trees possuem estrutura em forma de árvores.
* Random Forest pode ser utilizado tanto para classificação (RandomForestClassifier) quanto para Regressão (RandomForestRegressor);
* Os nós da árvore são criados a partir das variáveis do dataframe;

* **Vantagens**:
    * Não requer tanto data preprocessing;
    * Lida bem com COLUNAS categóricas e numéricas;
    * Apresenta bons resultados em diversos tipos de problema;
    * Estes modelos aprendem com os próprios erros e ajustam as árvores de modo a fazer melhores classificações;
        * Ensemble é a combinação de diferentes modelos preditivos;
        * Torna os algoritmos/resultados mais robustos e complexos, levando a um maior custo computacional que costuma ser acompanhando de melhores resultados.
    * Mais robusta que uma simples Decision Tree. **Porque?**
    * Controla automaticamente overfitting (**porque?**) e frequentemente produz modelos muito robustos e de alta-performance.
    * Pode ser utilizado como **Feature Selection**, pois gera a matriz de importância dos atributos (importance sample).
        * A soma das importâncias das variáveis soma 100;
    * Assim como as Decision Trees, esses modelos capturam facilmente padrões não-lineares presentes nos dados;
    * Não requer dados normalizados;
    * Lida bem com Missing Values;
    * Não requer suposições (assumptions) sobre a distribuição dos dados por causa da natureza não-paramétrica do algoritmo;

* **Desvantagens/Cuidados**
    * **Recomenda-se balancear o dataframe previamente**.

### Como funciona?

O algoritmo possui 4 passos:
1. Seleção aleatória de algumas features;
2. Seleção da feature mais adequada para a posição de nó raiz;
3. Geração dos nós filhos
4. Repete os passos acima até que se atinja a quantidade de árvores desejada.

**Observação**: Depois que o modelo é gerado, as previsões são feitas a partir de “votações” das várias árvores. A decisão mais votada é a resposta do algoritmo.

### Principais hiperparâmetros
* [Hyperparameter Tuning the Random Forest in Python](https://towardsdatascience.com/hyperparameter-tuning-the-random-forest-in-python-using-scikit-learn-28d2aa77dd74) - Explica os principais hiperparâmetros do Random Forest.

    * n_estimators: especifica o número de árvores. O valor default é 10, o que significa que 10 diferentes modelos de DecisionTrees serão ajustadas;
    * max_depth: especifica a profundidade máxima de cada árvore. O valor default é None, o que significa que cada árvore se expandirá até que cada folha esteja pura. Uma folha pura é aquela em que todos os dados da folha vêm da mesma classe;
    * min_samples_split: especifica o número mínimo de amostras necessárias para dividir um nó filho. O valor default é 2, o que significa que um nó filho deve ter pelo menos duas amostras antes de ser dividido;
    * min_samples_leaf: especifica o número mínimo de amostras necessárias para estar em um nó folha. O valor default é 1, o que significa que cada folha deve ter pelo menos 1 amostra para classificar.

# Exemplo
- A seguir, um dataframe (fake) para introduzirmos os principais conceitos relacionados ao Random Forest. Este dataset foi projetado para ensinar Random Forest e contém 2500 linhas e 15 colunas, representando variáveis comuns em problemas de classificação.

## Estrutura das Colunas:
* age: Idade do indivíduo (inteiro).
* income: Renda anual (numérica contínua, em reais).
* education_level: Nível de escolaridade (ensino_medio, graduacao, mestrado, doutorado).
* years_experience: Anos de experiência no trabalho (inteiro).
* credit_score: Pontuação de crédito (numérica contínua, variando entre 300 e 850).
* owns_car: Possui carro? (0 = não, 1 = sim).
* owns_home: Possui casa própria? (0 = não, 1 = sim).
* marital_status: Estado civil (solteiro, casado, divorciado).
* num_children: Número de filhos (inteiro).
* monthly_spending: Gastos médios mensais (numérica contínua, em reais).
* savings: Poupança total acumulada (numérica contínua, em reais).
* loan_amount: Valor do empréstimo solicitado (numérica contínua, em reais).
* loan_term: Prazo do empréstimo em anos (inteiro: 5, 10, 15, 20, 25, 30).
* defaulted_loan: A pessoa não conseguiu pagar o empréstimo? (0 = não, 1 = sim), variável-alvo.
* industry: Setor de trabalho (tecnologia, financas, saude, educacao, varejo).


```
import pandas as pd
import numpy as np

# Define uma seed para garantir a reprodutibilidade
np.random.seed(42)

# Gera um dataset sintético com 2500 linhas e 15 colunas
df = pd.DataFrame({
    'age': np.random.randint(18, 70, 2500),
    'income': np.random.normal(50000, 15000, 2500),
    'education_level': np.random.choice(['high_school', 'bachelor', 'master', 'phd'], 2500),
    'years_experience': np.random.randint(1, 40, 2500),
    'credit_score': np.random.normal(600, 100, 2500),
    'owns_car': np.random.choice([0, 1], 2500),  # variável binária
    'owns_home': np.random.choice([0, 1], 2500),  # variável binária
    'marital_status': np.random.choice(['single', 'married', 'divorced'], 2500),
    'num_children': np.random.randint(0, 5, 2500),
    'monthly_spending': np.random.normal(2000, 500, 2500),
    'savings': np.random.normal(20000, 10000, 2500),
    'loan_amount': np.random.normal(100000, 50000, 2500),
    'loan_term': np.random.choice([5, 10, 15, 20, 25, 30], 2500),
    'defaulted_loan': np.random.choice([0, 1], 2500, p=[0.8, 0.2]),  # variável-alvo
    'industry': np.random.choice(['tech', 'finance', 'healthcare', 'education', 'retail'], 2500)
})

# Ajusta quaisquer outliers ou valores irreais nas variáveis contínuas
df['income'] = df['income'].clip(lower=20000, upper=150000)
df['credit_score'] = df['credit_score'].clip(lower=300, upper=850)
df['monthly_spending'] = df['monthly_spending'].clip(lower=500, upper=10000)
df['savings'] = df['savings'].clip(lower=0)
df['loan_amount'] = df['loan_amount'].clip(lower=0)

# Exibe a estrutura do dataFrame gerado e salva para o usuário
df.head()
```

## Importar as bibliotecas básicas

```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

## Pré-processamento dos dados
* Vamos transformar as variáveis categóricas em variáveis numéricas usando `pd.get_dummies()`

```
# Converte variáveis categóricas em dummies
df = pd.get_dummies(df, columns=['education_level', 'marital_status', 'industry'], drop_first=True)
df.head()
```

```
# Define X (features) e y (variável-alvo)
X = df.drop('defaulted_loan', axis=1)
y = df['defaulted_loan'] # Variável target

# Divide os dados em conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Treinamento do modelo Random Forest
* Para treinar o modelo de Random Forest, usamos o `RandomForestClassifier`.

```
# Cria o modelo Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Treina o modelo com os dados de treino
rf_model.fit(X_train, y_train)

```

## Avaliação do Modelo
* Depois de treinar o modelo, avaliamos seu desempenho nos dados de teste.

```
# Faz previsões nos dados de teste
y_pred = rf_model.predict(X_test)

# Avalia o desempenho do modelo
print("Acurácia:", accuracy_score(y_test, y_pred))
print("\nMatriz de Confusão:\n", confusion_matrix(y_test, y_pred))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred))
```

# Importância das variáveis

```
# Calcula a importância das variáveis
importances = rf_model.feature_importances_

# Exibe as features mais importantes
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

print("\nImportância das Features:\n")
print(feature_importance.head(10))
```
