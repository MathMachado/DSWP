# Modelos de Regressão

Os métodos de regressão surgiram há mais de 2 séculos com Legendre (1805) e Gauss (1809), que exploram o método dos mínimos quadrados com o objetivo de prever órbitas ao redor do Sol.

Até meados dos anos 60 os modelos normais lineares foram utilizados na tentativa de descrever a maioria dos fenômenos aleatórios, onde o método dos mínimos quadrados eram utilizados para estimativa dos parâmetros. Estes modelos eram aplicados mesmo quando o fenômeno sob estudo não apresentava uma resposta para a qual fosse razoável a suposição de normalidade.

Em 1964, George Box e David Cox propuseram um procedimento de transformação a fim de alcançar a normalidade desejada. Esse método ficou conhecido como transformação de Box-Cox, o qual transformava o valor observado y (positivo) em:

$$
y(\lambda) =
\begin{cases}
\frac{y^\lambda - 1}{\lambda}, & \text{se } \lambda \neq 0 \\
\ln(y), & \text{se } \lambda = 0
\end{cases}
$$

em que $λ$ uma constante desconhecida.A ideia é encontrar um valor de $\lambda$ até encontrarmos a normalidade.

Portanto, o objetivo da transformação de Box-Cox, quando aplicado a um conjunto de dados não normais observados, é produzir aproximadamente a normalidade, variabilidade constante e também a linearidade.

$E(Z) = η = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_p x_p$.

No entanto, observou-se que todas as suposições de normalidadade ocorrem para  mais de um valor de $\lambda$, como verificaram Box e Draper (1987). Ou seja, a linearidade $E(Z) = η = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_p x_p$ raramente ocorre para um único valor de $λ$.

## Seleção de modelos
A seleção de modelos de regressão é o processo mais importante de toda a pesquisa em modelagem estatística.

Isso consiste em determinar um modelo que seja o mais simples possível e que descreva bem o processo gerador dos valores observados que surgem em diversas áreas do conhecimento tais como:
* Agricultura, Ecologia, Zoologia;
* Demografia e Economia;
* Engenharias e Geologia;
* Ciências Biológicas e médicas;
* Ciências políticas, sociologia;
* etc.,

Em 1972, Nelder e Wedderburn mostraram que um conjunto de técnicas estatísticas comumente estudadas separadamente, podem ser formuladas, de uma maneira unificada, como uma classe de modelos de regressão.

A essa teoria unificada de modelagem estatística deu-se o nome de Modelos Lineares Generalizados (GLM), que são uma extensão dos modelos lineares clássicos de regressão.

## Os Modelos Lineares Generalizados (GLMs)

Abaixo, consolidei as informações do material [cite: 1] em um guia teórico estruturado, eliminando as instruções de software e focando na **morfologia matemática dos MLGs**.

---

## 1. A Família Exponencial Linear
A base teórica de um MLG é a suposição de que a variável resposta $Y_i$ pertence à família exponencial linear, cuja forma canônica é definida por[cite: 34, 35, 37]:
$$f(y_i;\theta_i,\phi)=\exp[\phi\{y_i\theta_i-b(\theta_i)\}+c(y_i,\phi)]$$

* **Parâmetros:** $\theta_i$ é o parâmetro canônico e $\phi$ é o parâmetro de precisão (onde $\phi^{-1}$ é a dispersão)[cite: 35, 39].
* **Momentos via b(θ):** As propriedades de média e variância derivam diretamente da função cumulante $b(\theta)$[cite: 39]:
    * $E(Y_i) = \mu_i = b'(\theta_i)$[cite: 38, 39].
    * $Var(Y_i) = \phi^{-1} V(\mu_i)$, onde $V(\mu_i) = b''(\theta_i) = d\mu_i/d\theta_i$ é a **função de variância**[cite: 39].
* **Caracterização:** A função de variância é o que caracteriza a distribuição; dada uma $V(\mu)$, tem-se uma classe de distribuições correspondentes[cite: 39, 40].



---

## 2. Componentes Estruturais do Modelo
O MLG é composto pela distribuição acima e pela **parte sistemática**, que estabelece a relação entre a média e os preditores[cite: 48, 49]:
1.  **Preditor Linear ($\eta_i$):** $\eta_i = x_i^\top \beta$, onde $\beta$ é o vetor de parâmetros desconhecidos[cite: 51].
2.  **Função de Ligação ($g$):** Uma função monótona e diferenciável tal que $g(\mu_i) = \eta_i$[cite: 49, 51].
3.  **Ligação Canônica:** Ocorre quando $g(\mu_i) = \theta_i$[cite: 91, 96, 101]. Teoricamente, ela garante a concavidade da função de log-verossimilhança $L(\beta)$, assegurando a unicidade da estimativa de máxima verossimilhança[cite: 104, 105].

---

## 3. Teoria da Estimação e Função Escore
A estimação em MLGs utiliza as **Condições de Regularidade**, que relacionam a primeira e a segunda derivada do logaritmo da função de densidade[cite: 37, 38]:
$$E \left\{ \frac{\partial \log f(Y_i)}{\partial \theta_i} \right\} = 0 \quad \text{e} \quad E \left[ \frac{\partial^2 \log f(Y_i)}{\partial \theta_i^2} \right] = -E \left[ \left\{ \frac{\partial \log f(Y_i)}{\partial \theta_i} \right\}^2 \right]$$

* **Função Escore ($U_\beta$):** Expressa matricialmente como $U_\beta = \phi X^\top W^{1/2} V^{-1/2} (y - \mu)$, onde $W$ é a matriz de pesos diagonal com elementos $\omega_i = (d\mu_i/d\eta_i)^2/V_i$[cite: 346, 347, 348, 349].
* **Informação de Fisher ($K_{\beta\beta}$):** A submatriz de informação esperada é $K_{\beta\beta} = \phi X^\top WX$[cite: 358].
* **Ortogonalidade:** Um resultado teórico vital é que $\beta$ e $\phi$ são ortogonais ($K_{\beta\phi} = 0$), o que implica independência assintótica entre as estimativas dos coeficientes e do parâmetro de dispersão[cite: 368, 369].



---

## 4. Algoritmo de Mínimos Quadrados Reponderados (IWLS)
Para obter $\hat{\beta}$, utiliza-se o método de **Escore de Fisher**, que resulta em um processo iterativo de mínimos quadrados reponderados[cite: 412, 413, 415]:
$$\beta^{(m+1)} = (X^\top W^{(m)} X)^{-1} X^\top W^{(m)} z^{(m)}$$
* **Variável Dependente Modificada ($z$):** $z = \eta + W^{-1/2} V^{-1/2} (y - \mu)$[cite: 418].
* **Natureza do Algoritmo:** Note que o lado direito da equação não depende de $\phi$, permitindo estimar $\beta$ sem conhecer o parâmetro de dispersão[cite: 425].

---

## 5. Inferência, Desvio e Assintótica
A avaliação do modelo baseia-se na **Função Desvio (Deviance)**, uma medida de distância entre o logaritmo da verossimilhança do modelo saturado e do modelo sob investigação[cite: 196, 197, 198]:
$$D^*(y;\hat{\mu}) = 2\{L(y;y) - L(\hat{\mu};y)\}$$

* **Resultados Assintóticos:**
    * **Consistência:** Sob condições de regularidade, $\hat{\beta}$ segue assintoticamente uma distribuição normal $N_p(\beta, K_{\beta\beta}^{-1})$[cite: 462].
    * **Distribuição do Desvio:** Para dispersão pequena ($\phi \to \infty$), o desvio escalonado $D^*(y;\hat{\mu})$ segue aproximadamente uma distribuição $\chi^2_{n-p}$[cite: 270, 271, 272].
    * **Normalidade Assintótica da Resposta:** À medida que a precisão $\phi$ aumenta, a variável padronizada converge em distribuição: $\sqrt{\phi}(Y-\mu) \rightarrow_d N(0, V(\mu))$[cite: 42].



---

### Resumo das Principais Distribuições (Foco Teórico) [cite: 90]
| Distribuição | $b(\theta)$ | $\theta$ | $V(\mu)$ |
| :--- | :--- | :--- | :--- |
| **Normal** | $\theta^2/2$ | $\mu$ | $1$ |
| **Poisson** | $e^\theta$ | $\log(\mu)$ | $\mu$ |
| **Binomial** | $\log(1+e^\theta)$ | $\log\{\mu/(1-\mu)\}$ | $\mu(1-\mu)$ |
| **Gama** | $-\log(-\theta)$ | $-1/\mu$ | $\mu^2$ |
| **N. Inversa** | $-\sqrt{-2\theta}$ | $-1/2\mu^2$ | $\mu^3$ |

### Diagnóstico Essencial: Superdispersão (Overdispersion)

Em um GLM Poisson, a teoria diz que a **Média deve ser igual à Variância**.
* Se a Variância for muito maior que a Média, seu modelo está "mentindo" sobre a precisão (os p-valores ficam artificialmente baixos).
* **Como checar?** Divida a `Pearson chi2` pelos `Df Residuals`. Se o resultado for muito maior que 1.2, você tem superdispersão e deveria usar um modelo **Binomial Negativa**.

### O que falta para virar GLMM?
Neste exemplo, assumimos que cada vendedor é independente. Mas e se esses vendedores estivessem agrupados em **10 Filiais diferentes**?
* Vendedores da mesma filial compartilham o mesmo gerente e clima regional.
* O **GLM** ignoraria isso e poderia dar resultados enviesados.
* O **GLMM** adicionaria um "Efeito Aleatório de Filial" para absorver essa dependência.

# GLM vs GLMM

Para explicar a diferença fundamental entre os Modelos Lineares Generalizados (**GLM**) e os Modelos Lineares Generalizados Mistos (**GLMM**), é preciso focar na estrutura do **preditor linear** e na suposição de **independência** das observações.

O material fornecido define os **MLGs** (GLMs) no Capítulo 2, enquanto introduz os **MLGMs** (GLMMs) como uma extensão importante na introdução teórica[cite: 11, 23].

---

## 1. O Preditor Linear: Fixo vs. Misto

A principal distinção reside na composição do preditor linear ($\eta$):

* **GLM (Modelos Lineares Generalizados):** O preditor linear é composto estritamente por **efeitos fixos**[cite: 51].
    * Equação: $\eta_i = x_i^\top \beta$[cite: 51].
    * Aqui, $\beta$ representa parâmetros constantes que descrevem a relação média para toda a população sob estudo[cite: 51].
* **GLMM (Modelos Lineares Generalizados Mistos):** O preditor linear é **misto**, pois admite a inclusão de **efeitos aleatórios** além dos efeitos fixos[cite: 23].
    * Equação (implícita): $\eta_i = x_i^\top \beta + z_i^\top u$.
    * O termo $u$ representa efeitos aleatórios, geralmente assumidos como seguidores de uma distribuição normal[cite: 23].



---

## 2. Dependência vs. Independência dos Dados

Essa diferença estrutural altera a suposição fundamental sobre os dados:

* **GLM:** Assume que as variáveis aleatórias $Y_1, \dots, Y_n$ são **independentes**[cite: 34]. Ele é adequado para dados onde não há agrupamento ou correlação entre as observações (ex: uma amostra aleatória simples da população)[cite: 34].
* **GLMM:** É projetado especificamente para situações onde existe **correlação** ou dependência entre as observações[cite: 23]. Isso ocorre frequentemente em:
    * **Dados Hierárquicos:** Onde o preditor linear pode ter efeitos aleatórios não gaussianos em diferentes níveis[cite: 24].
    * **Medidas Repetidas:** Quando o mesmo indivíduo é medido várias vezes.
    * **Estudos de Agrupamento (Clustering):** Onde indivíduos dentro de um mesmo grupo (ex: mesma escola ou hospital) tendem a ser mais parecidos entre si do que com indivíduos de outros grupos.

---

## 3. Resumo Comparativo para Concursos

| Característica | GLM (MLG) | GLMM (MLGM) |
| :--- | :--- | :--- |
| **Preditor Linear** | Apenas efeitos fixos ($\beta$)[cite: 51]. | Efeitos fixos ($\beta$) + Aleatórios ($u$)[cite: 23]. |
| **Independência** | Exige observações independentes[cite: 34]. | Permite observações correlacionadas/dependentes[cite: 23]. |
| **Distribuição** | Resposta na Família Exponencial[cite: 35]. | Resposta na Família Exponencial com efeitos aleatórios (geralmente gaussianos)[cite: 23]. |
| **Complexidade** | Estimação via IWLS (Newton-Raphson)[cite: 405, 412]. | Exige métodos de integração numérica ou aproximações mais complexas devido aos efeitos aleatórios[cite: 10, 23]. |



Em suma, enquanto o **GLM** estende o modelo linear clássico para distribuições não-normais (como Poisson ou Binomial) [cite: 13, 14], o **GLMM** leva essa extensão um passo adiante, permitindo modelar a estrutura de correlação dos dados através de componentes aleatórios no preditor[cite: 23, 24].

# GLMM (Modelo Linear Generalizado Misto)

O GLMM é o "canivete suíço" do cientista de dados: ele combina a flexibilidade dos **Modelos Lineares Generalizados (GLM)** — que lidam com distribuições não-normais (Poisson, Binomial) — com a capacidade dos **Modelos Mistos** de lidar com dependência de dados (agrupamentos).

## 1. A Estrutura do Modelo
A forma matemática simplificada de um GLMM pode ser expressa como:

$$g(\mu) = X\beta + Zu + \epsilon$$

Onde:
* $g(\cdot)$: É a **função de ligação** (ex: *logit* para binários, *log* para contagens).
* $X\beta$: Representa os **Efeitos Fixos** (os preditores que queremos estimar).
* $Zu$: Representa os **Efeitos Aleatórios** (a estrutura de erro/agrupamento).
* $\epsilon$: O resíduo.

## 2. Efeitos Fixos vs. Aleatórios

### **Efeitos Fixos (Fixed Effects)**
São os níveis de uma variável que estamos interessados em estudar diretamente. Eles são constantes entre os grupos.
* **Exemplo:** O efeito de um novo medicamento vs. placebo.

### **Efeitos Aleatórios (Random Effects)**
São níveis de uma variável que representam uma amostra de uma população maior. Eles introduzem uma correlação entre as observações do mesmo grupo.
* **Exemplo:** O "Efeito do Hospital". Se você coleta dados de 10 hospitais, os pacientes dentro de cada hospital são mais parecidos entre si do que com pacientes de outros locais. O hospital é o seu *random effect*.

## 3. Exemplos Práticos

| Contexto | Variável Resposta (Y) | Efeito Fixo (X) | Efeito Aleatório (Z) |
| :--- | :--- | :--- | :--- |
| **Ecologia** | Presença/Ausência de espécie | Temperatura e Umidade | Local da coleta (Sítio) |
| **Educação** | Nota na prova | Método de ensino | Sala de aula / Aluno (Medidas repetidas) |
| **Medicina** | Número de crises epilépticas | Dosagem do remédio | Paciente (ID) |

## 4. Ajuste e Diagnóstico

### **Ajuste (Fitting)**
O ajuste geralmente é feito via **Máxima Verossimilhança Restrita (REML)** ou **Integração Numérica (Laplace/GHQ)**. No R, a biblioteca padrão é a `lme4` (função `glmer`) ou a `glmmTMB`.

### **Diagnóstico e Validação**
Validar um GLMM é mais desafiador que um modelo linear simples. Não olhamos apenas para o $R^2$.
1.  **Resíduos de DHARMa:** Para GLMMs, resíduos brutos são enganosos. Usamos simulações (pacote DHARMa no R) para verificar se há sobre-dispersão ou outliers.
2.  **Convergência:** O modelo "chegou lá"? Erros de *Singularity* indicam que sua estrutura de efeitos aleatórios é complexa demais para os dados disponíveis.
3.  **ICC (Intraclass Correlation Coefficient):** Mede quanto da variação total é explicada pelo agrupamento.

## 5. Interpretação para Dados Agrupados

Ao interpretar um GLMM, você deve ter em mente que:
* **Efeito Marginal:** É a média da população.
* **Efeito Condicional:** É o efeito para aquele grupo/indivíduo específico.

Se você tem um coeficiente de $\beta = 0.5$ para um tratamento em um modelo Logístico (Binomial), isso significa que, **dentro de um mesmo grupo (ex: mesmo hospital)**, o aumento de uma unidade no tratamento aumenta o *log-odds* da resposta em 0.5.

## Resumo da Ópera
O GLMM é essencial quando seus dados não são independentes. Ignorar os efeitos aleatórios (tratar tudo como fixo) infla o erro tipo I — ou seja, você achará que seus resultados são estatisticamente significativos quando, na verdade, é apenas o efeito do agrupamento.

# Exemplo
Imagine que estamos avaliando o impacto de um **Método de Ensino** (Tradicional vs. Inovador) na probabilidade de um aluno ser **aprovado** (0 ou 1). Como alunos da mesma sala tendem a ter um desempenho similar (devido ao mesmo professor ou ambiente), os dados são **dependentes**.

Em Python, a biblioteca mais robusta e utilizada para **GLMM** (Modelos Lineares Generalizados Mistos) é a `statsmodels`. Embora o ecossistema Python seja historicamente mais focado em Machine Learning, o módulo `MixedLM` e as extensões de fórmulas facilitam muito essa tarefa estatística.

Para o nosso exemplo de **Educação** (Aprovação ~ Método + Efeito Aleatório por Sala), utilizaremos uma abordagem de **Equações de Estimativa Generalizadas (GEE)** ou o **Binomial Mixed Model** via `statsmodels`.

---

## Implementação com `statsmodels`

O Python exige que a gente trate a variável de agrupamento (ID da sala) explicitamente no argumento `groups`.



### Gerando o Dataset Sintético
Aqui está o código para gerar o DataFrame dados_escola com as colunas que precisamos:

````python
import pandas as pd
import numpy as np

# Para garantir que os resultados sejam reproduzíveis
np.random.seed(42)

# Configurações do dataset
n_salas = 10           # 10 salas de aula
alunos_por_sala = 30   # 30 alunos em cada
n_total = n_salas * alunos_por_sala

# 1. Criando os IDs das salas e os métodos
sala_id = np.repeat(np.arange(1, n_salas + 1), alunos_por_sala)
# Atribuindo o método de forma aleatória por sala (cluster)
metodos_escolhidos = np.random.choice(['Tradicional', 'Inovador'], size=n_salas)
metodo = np.repeat(metodos_escolhidos, alunos_por_sala)

# 2. Gerando o efeito aleatório da sala (Random Intercept)
# Algumas salas têm uma "base" de aprovação maior que outras
efeito_sala = np.repeat(np.random.normal(0, 1.5, size=n_salas), alunos_por_sala)

# 3. Gerando a probabilidade de aprovação
# Beta0 (intercepto) = -0.5
# Beta1 (efeito do método inovador) = 1.2
metodo_num = np.where(metodo == 'Inovador', 1, 0)
log_odds = -0.5 + (1.2 * metodo_num) + efeito_sala + np.random.normal(0, 0.5, n_total)

# Convertendo log-odds para probabilidade (função logística)
probabilidade = 1 / (1 + np.exp(-log_odds))

# 4. Gerando a variável resposta binária (Aprovado: 0 ou 1)
aprovado = np.random.binomial(1, probabilidade)

# Montando o DataFrame
dados_escola = pd.DataFrame({
    'sala_id': sala_id,
    'metodo': metodo,
    'aprovado': aprovado
})

print(dados_escola.head())
print(f"\nTaxa de aprovação por método:\n{dados_escola.groupby('metodo')['aprovado'].mean()}")
````

Para este exemplo de **GLM (Modelo Linear Generalizado)**, vamos focar no modelo **Logístico Binomial**. Diferente do modelo anterior (`MixedLM`), aqui tratamos a variável `aprovado` como uma probabilidade entre 0 e 1, usando a função de ligação *logit*.

Vamos assumir que rodamos o seguinte código no Python:

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

# Ajustando o GLM Binomial
modelo_glm = smf.glm("aprovado ~ metodo",
                     data=dados_escola,
                     family=sm.families.Binomial()).fit()

print(modelo_glm.summary())
```

### Interpretação dos Parâmetros

No GLM Logístico, os coeficientes (**coef**) não são probabilidades diretas, mas sim **Log-Odds** (logaritmo da razão de chances).

#### **A. Intercept (0.5108)**
Este é o log-odds de aprovação para o grupo de referência (Método Tradicional).
* Para entender o que isso significa em probabilidade:

    $$\text{Probabilidade} = \frac{1}{1 + e^{-0.5108}} \approx 0.625 \text{ ou 62.5%}$$

* **Interpretação:** A taxa de aprovação esperada no método tradicional é de 62,5%.

#### **B. metodo[T.Inovador] (0.8473)**
Este é o efeito do método inovador em relação ao tradicional. Como o valor é positivo e o **P-valor (0.001)** é menor que 0.05, o efeito é estatisticamente significativo.
* **Cálculo da Razão de Chances (Odds Ratio):**
    $\text{Odds Ratio} = e^{0.8473} \approx 2.33$
* **Interpretação:** Alunos que utilizam o **Método Inovador** têm **2,33 vezes mais chances** (odds) de serem aprovados do que os alunos do método tradicional.



---

### Onde entra o "Misto" (GLMM)?

Se você adicionar o efeito aleatório da sala de aula a este GLM, você terá um **GLMM**.
* **GLM (este aqui):** Diz que o Método Inovador é bom para a média de todos os alunos.
* **GLMM:** Diria se o Método Inovador continua sendo bom **mesmo quando descontamos o fato de que algumas salas têm professores melhores que outras**.

Basicamente, o GLMM "limpa" a influência da sala de aula para isolar o verdadeiro poder do método de ensino. Ficou clara a diferença entre interpretar o coeficiente puro e a razão de chances (Odds Ratio)?

# Nível mestre: GLMM
Agora vamos ao "nível mestre". No **GLMM (Modelo Linear Generalizado Misto)**, combinamos a função de ligação **Logística (Binomial)** com o **Efeito Aleatório da sala**.

Como o Python (`statsmodels`) tem uma saída de GLMM um pouco mais técnica, vou apresentar um output padrão baseado no que você encontraria ao rodar um modelo robusto (como no `Pymer4` ou `lme4`), que é o padrão ouro na academia e indústria.

Imagine que rodamos: `aprovado ~ metodo + (1 | sala_id)` com família `Binomial`.

---

### 1. Output Hipotético do GLMM (Logístico Misto)

```text
Linear mixed model fit by maximum likelihood (Adaptive Gauss-Hermite Quadrature)
Formula: aprovado ~ metodo + (1 | sala_id)
Family: Binomial ( link = logit )

Random effects:
 Groups   Name        Variance Std.Dev.
 sala_id  (Intercept) 0.640    0.800   
Number of obs: 300, groups: sala_id, 10

Fixed effects:
                   Estimate  Std. Error  z value  Pr(>|z|)    
(Intercept)         0.450     0.320       1.406    0.1597    
metodo[T.Inovador]  1.100     0.410       2.682    0.0073 ** ---
Signif. codes:  0 '***' 0.001 '**' 0.01 '*' 0.05
```

---

### 2. Interpretação dos Parâmetros

Aqui a interpretação se divide em duas partes: o que é comum a todos (Fixo) e o que varia entre grupos (Aleatório).

#### **A. Efeitos Aleatórios (Random Effects)**
* **Variance (0.640):** Esta é a variabilidade entre as salas de aula.
* **Interpretação:** Como o valor não é zero, isso confirma que **a sala de aula importa**. Algumas salas têm uma tendência intrínseca maior de aprovação (talvez pelo clima escolar) do que outras. O GLMM "limpa" essa interferência para que ela não vicie o efeito do método.



#### **B. Efeitos Fixos (Fixed Effects)**
* **Intercept (0.450):** É o log-odds de aprovação no método de referência (Tradicional) para uma sala "média" (onde o efeito aleatório é zero).
* **metodo[T.Inovador] (1.100):** Este é o coeficiente principal.
    * **P-valor (0.0073):** É significativo! O método inovador realmente faz diferença, mesmo após controlarmos a variação entre as salas.
    * **Cálculo do Odds Ratio:** $e^{1.100} \approx 3.00$.
    * **Interpretação Condicional:** Para um aluno **dentro da mesma sala**, mudar do método tradicional para o inovador **triplica (3x)** as chances de aprovação.

---

### 3. A Grande Diferença: GLM vs. GLMM

Se você olhar o coeficiente do método no **GLM** (exemplo anterior, era ~0.84) e comparar com o **GLMM** (este agora, 1.10), você notará que o coeficiente do GLMM costuma ser **maior**.

**Por que?**
O GLM tenta explicar a média de toda a escola (efeito marginal). O GLMM explica o que acontece com o indivíduo quando você remove o "ruído" da sala de aula (efeito condicional). Em modelos logísticos, o efeito "dentro do grupo" costuma ser mais puro e forte.



### 4. Diagnóstico Final
Para validar este modelo, você verificaria:
1.  **ICC (Coeficiente de Correlação Intraclasse):** Calcula-se como $\frac{Var_{Grupo}}{Var_{Grupo} + (\pi^2 / 3)}$. No nosso caso, diria qual a porcentagem da variância da aprovação que é culpa exclusiva da sala.
2.  **Convergência:** Se o otimizador não "sofreu" para encontrar esses números.
