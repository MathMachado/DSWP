```
# Importa Bibliotecas iniciais
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Importa DataFrame
url = 'https://raw.githubusercontent.com/Mvital74/DSWP/master/Dataframes/FIFA.csv'
fifa18 = pd.read_csv(url, index_col='ID')
fifa18.head()
```

```
fifa18_tidy = fifa18.drop(columns=['Name', 'Club','Unnamed: 0', 'Photo', 'Nationality','Height', 'Weight', 'Flag', 'Club Logo', 'Preferred Foot', 'Body Type', 'Real Face', 'Jersey Number', 'Joined', 'Loaned From', 'Contract Valid Until', 'Release Clause'], axis=1)
fifa18_tidy.head()
```

```
fifa18_tidy = fifa18_tidy.drop(list(fifa18_tidy)[11:37], axis=1)
```

```
#Transformar para lower
def transformacao_lower(df):
    # Primeira transformação: Aplicar lower() nos nomes das COLUNAS:
    df.columns = [col.lower() for col in df.columns]
```

```
transformacao_lower(fifa18_tidy)
```

```
fifa18_gk = fifa18_tidy[fifa18_tidy['position'] == 'GK']
fifa18_gk.head()
```

```
fifa18_pl = fifa18_tidy[fifa18_tidy['position'] != 'GK']
fifa18_pl.head()
```

```
fifa18_pl.loc[:,'value'] = fifa18_pl.loc[:,'value'].str.replace('€','')
fifa18_pl.loc[:,'wage'] = fifa18_pl.loc[:,'wage'].str.replace('€','')
```

```
fifa18_pl.head()
```

```
fifa18_pl['wage'] = fifa18_pl['wage'].map(lambda x: int(x.replace('K',''))*1000)
```

```
fifa18_pl.head()
```

```
def money_num(money):
  if money[-1] == 'M':
    return float(money.replace('M',''))*1000000
  else:
    return float(money.replace('K',''))*1000
```

```
fifa18_pl['value'] = fifa18_pl['value'].apply(money_num)
fifa18_pl.head()
```

```
fifa18_pl.info()
```

```
fifa18_na=fifa18_pl.dropna()
```
